# Name of the max valued product

```python
chipo.info()
```

```python
chipo['item_price'] = chipo['item_price'].str[1:].astype(float)
```

```python
chipo.head()
```

```python
chipo[chipo['item_price'] == chipo['item_price'].max()]
```

```python

chipo[chipo['item_name'] == 'Chicken Bowl']
```

```python
# Sacar precio unitario
chipo['unitary_price'] = chipo['item_price'] / chipo['quantity']
chipo.head()
```

```python
chipo2 = chipo[chipo['unitary_price'] == chipo['unitary_price'].max()]
chipo2[['item_name', 'unitary_price']].drop_duplicates()
```

### Step 4. How many products cost more than $10.00?

```python
products = chipo[chipo['item_price'] > 10]
products['item_name'].nunique()
```

```python
products = chipo[chipo['unitary_price'] > 10]
products['item_name'].nunique()
```

```python
chipo[chipo['unitary_price'] > 10.00].drop_duplicates(['item_name', 'unitary_price'])
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_name', 'item_price']].sort_values(by='item_price', ascending=False).drop_duplicates()
```
