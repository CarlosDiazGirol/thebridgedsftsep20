```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(123)
x = np.random.chisquare(5, 4000)
plt.hist(x, bins = 100)
plt.show()
```

```python
x_log_1 = np.log(x + 1)
plt.hist(x_log_1, bins = 100)
plt.show()
```

```python
x_log_1_st = (x_log_1 - x_log_1.mean()) / x_log_1.std()
plt.hist(x_log_1_st, bins = 100)
plt.show()
```

```python
x_log = np.log(x)
plt.hist(x_log, bins = 100)
plt.show()
```

```python
np.random.seed(123)
norm = np.random.normal(10, 1, 4000)
plt.hist(norm, bins = 100)
plt.show()
```

```python
from scipy.stats import boxcox

x_bc = boxcox(x)
plt.hist(x_bc, bins = 100)
plt.show()
```

```python
x_bc = boxcox(x)
x_bc
```

```python
from scipy.stats import shapiro

print(shapiro(x))
print(shapiro(norm))
print(shapiro(x_log_1))
print(shapiro(x_log))
print(shapiro(x_bc))
```

```python
np.random.seed(123)
tr_1 = np.random.triangular(0,50, 100, 4000)
tr_2 = np.random.triangular(0,0, 100, 4000)
plt.hist(tr_1, bins = 100)
plt.show()
print(shapiro(tr_1))
print(shapiro(tr_2))
```

```python
plt.hist(tr_2, bins = 100)
plt.show()
```

```python
tr_1_log = np.log(tr_1)
plt.hist(tr_1_log, bins = 100)
plt.show()
```

```python
log_norm = np.random.lognormal(0, 1, 4000)
plt.hist(log_norm, bins = 100)
plt.show()
```

```python
log_norm_log_1 = np.log(log_norm + 1)
log_norm_log = np.log(log_norm)
log_norm_bc = boxcox(log_norm)[0]
plt.hist(log_norm_bc, bins = 100)
plt.show()
print(shapiro(log_norm))
print(shapiro(log_norm_log))
print(shapiro(log_norm_log_1))
print(shapiro(log_norm_bc))
```
