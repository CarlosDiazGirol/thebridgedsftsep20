### Birthdays data

````python
births = pd.read_csv('data/births.csv')
births['decade'] = 10*(births['year']//10)
births.head()
````

````python
births.pivot_table(values='births',
                   index='decade',
                   columns='gender',
                   aggfunc=np.sum)
````

````python
df = births.groupby(['decade', 'gender']).sum()[['births']]
df.reset_index()
````

## Melt

````python
import pandas as pd

df_wide = pd.DataFrame(
  {"student": ["Andy", "Bernie", "Cindy", "Deb"],
   "school":  ["Z", "Y", "Z", "Y"],
   "english": [10, 100, 1000, 10000],  # eng grades
   "math":    [20, 200, 2000, 20000],  # math grades
   "physics": [30, 300, 3000, 30000]   # physics grades
  }
)
df_wide
````

````python
df_wide.melt(id_vars=["student", "school"],
            var_name="Columnas",
            value_name="Valores")
````

````python
df_wide.head()
````

````python
df_wide.melt(id_vars="student",
            value_vars=["english", "math"],
            var_name="Columnas",
            value_name="Valores")
````

````python
df_wide.melt(id_vars="student",
            var_name="Columnas",
            value_name="Valores")
````

## Pivot vs Melt
```Python
df.pivot_table(values, index, columns)
```

````python
df = df_wide.melt(id_vars="student",
            var_name="Columnas",
            value_name="Valores")

df.pivot_table(values = "Valores",
              index="student",
              columns="Columnas",
              aggfunc='sum')
````

````python
df.pivot(values = "Valores",
              index="student",
              columns="Columnas")
````
